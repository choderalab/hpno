```python
import torch
import dgl
%load_ext autoreload
%autoreload 2
import hpnotiq as hq
```

```python
ds = hq.data.esol()
ds_tr, ds_te = hq.data.utils.split(ds, [4, 1])
ds_tr = hq.data.utils.batch(ds_tr, 32)
ds_te = hq.data.utils.batch(ds_te, len(ds_te))
```

```python
net = torch.nn.Sequential(
    hq.models.input_layer.InputLayer(),
    hq.models.hmp.HierarchicalMessagePassing(),
    hq.models.output_layer.Pool(),
)
```

```python
optimizer = torch.optim.Adam(net.parameters(), 1e-3)
```

```python
loss_fn = torch.nn.functional.mse_loss
```

```python
for _ in range(100):
    for g, y in ds_tr:
        optimizer.zero_grad()
        y_hat = net(g).nodes['g'].data['y_hat'].flatten()
        loss = loss_fn(
            y.flatten(),
            y_hat
        )
        loss.backward()
        print(loss)
        optimizer.step()
```

```python
g, y = ds_te[0]
y_hat = net(g).nodes['g'].data['y_hat'].flatten()
y = y.flatten()
```
