```python
ds = hq.data.lipophilicity()
ds_tr, ds_te = hq.data.utils.split(ds, [4, 1])
ds_te = hq.data.utils.batch(ds_te, len(ds_te))
```

```python
net = torch.nn.Sequential(
    hq.models.input_layer.InputLayer(),
    hq.models.hmp.HierarchicalMessagePassing(),
    hq.models.output_layer.MultiPool(),
)
```

```python
# pinot
net = pinot.Net(
    pinot.representation.Sequential(
        pinot.representation.dgl_legacy.gn(),
        [32, 'tanh', 32, 'tanh', 32, 'tanh']),
    output_regressor=pinot.regressors.ExactGaussianProcessRegressor)
```

```python
optimizer = torch.optim.Adam(net.parameters(), 1e-3)
```

```python
loss_fn = torch.nn.functional.mse_loss
```

```python
for _ in range(100):
    for g, y in ds_tr:
        optimizer.zero_grad()
        loss = net.loss(g, y)
        loss.backward()
        print(loss)
        optimizer.step()
```

```python
g, y = ds_te[0]
y_hat = net(g).nodes['g'].data['y_hat'].flatten()
y = y.flatten()
```
